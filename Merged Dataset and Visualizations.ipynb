```python
#This function cleans the data from China csv file, drops irrelevant columns, separates composite values
#It returns chinese companies dataframe ready for combination.
#filename = "China.csv"
def load_ch_comapanies(filename):
    ch_companies = pd.read_csv(filename)
    ch_companies.drop(ch_companies.index[[128,129]], inplace = True)
    ch_companies = ch_companies.drop(ch_companies.columns[[3, 4, 5,14,15,16,17,18,24]], axis=1)
    ch_companies['markets'], ch_companies['other_markets'] = ch_companies['markets'].str.split(', ', 1).str
    if len(ch_companies['locations'].str.split(',')) > 1:
        ch_companies['locations'],ch_companies['other_locations'] = ch_companies['locations'].str.split(', ', 1).str
    ch_companies.drop(['id','community_profile','quality','high_concept','follower_count','company_url','created_at','updated_at','crunchbase_url','company_type','status','other_markets','other_locations', 'quality', 'product_desc','follower_count', 'company_url', 'created_at','updated_at', 'crunchbase_url', 'company_size',
    'status', 'other_markets', 'other_locations'],axis = 1,inplace = True)
    
    ch_companies['Startup_Country'] = 'China'
    ch_companies.rename(index=str, columns = {'name':'Startup_Name',
                                            'markets':'Startup_Industry_Vertical',
                                            'locations':'Startup_Location'},inplace = True)
    return ch_companies
#load_ch_comapanies(filename)
```

```python
#This function extract the company information from the websraping and crunchbase combined csv file.
#It returns a company datafram ready for combination.
#filename = "Merged_WebScrapping_CrunchBase_Investment.csv"
def load_web_cb_company(filename):
    df = pd.read_csv(filename)
    companies = df.drop(['Investor_Industry_Vertical','Investor_Location', 'Investor_Name'],axis = 1)
    companies = companies[companies.Startup_Country!='2008-02']
    return companies
#load_web_cb_company(filename)
```

```python
#This function extract the company information from csv from the api file.
#It returns dataframes of investors and companies.
#filename = "apidata2.csv"
def read_api_csv(filename):
    df = pd.read_csv(filename)
    df = df[pd.notnull(df['name'])]
    
    investors = df.loc[df['primary_role'] == 'investor']
    companies = df.loc[df['primary_role'] == 'company']
    
    investors = investors[['name', 'city_name']]
    companies = companies[['name', 'city_name','country_code']]
    
    investors.rename(index=str, columns = {'name':'Investor_Name','city_name':'Investor_Location'},inplace = True)
    companies.rename(index=str, columns = {'name':'Startup_Name','city_name':'Startup_Location','country_code':'Startup_Country'},inplace = True)
    
    return investors, companies
#i,c = read_api_csv(filename)
#i
#c
```

```python
#This function extract the investor information from the websraping and crunchbase combined csv file.
#filename = "Merged_WebScrapping_CrunchBase_Investment.csv"
def load_web_cb_investor(filename):
    df = pd.read_csv(filename)
    investors = df[pd.notnull(df['Investor_Name'])]
    investors = df.drop(['Startup_Country','Startup_Industry_Vertical', 'Startup_Location', 'Startup_Name'],axis = 1)
    
    investors.fillna({'Investor_Industry_Vertical':'Venture Capital',
                 'Investor_Location':'International'},inplace = True)
    return investors
#load_web_cb_investor("Merged_WebScrapping_CrunchBase_Investment.csv")
```

```python
#This function combines all the companies information we have from multiple sources
#It takes filenames and returns integrated dataframe.
def join_companies(chinese_file,merged_file,api_csv_file):
    ch_companies = load_ch_comapanies(chinese_file)
    merged_companies = load_web_cb_company(merged_file)
    merged_investors = load_web_cb_investor(merged_file)
    api_investors,api_companies = read_api_csv(api_csv_file)
    
    all_companies = merged_companies.append(ch_companies, sort = True)
    all_companies = all_companies.append(api_companies, sort = True)
    all_companies.drop_duplicates(['Startup_Name'], keep='first', inplace=True)
    
    all_investors = merged_investors.append(api_investors, sort = True)
    all_investors.drop_duplicates(['Investor_Name'], keep='first', inplace=True)
    all_investors.fillna({'Investor_Industry_Vertical':'Venture Capital',
                 'Investor_Location':'International'},inplace = True)
   
    return all_companies, all_investors
```

```python
def Clean_file(scrapping_FilePath):
    #Python Group-6_Data Focused Python_12thOct2018\StartupsInvestorsWebScrapping.csv
    data = pd.read_csv(scrapping_FilePath, header=1)
    # Renaming the column names to ensure integrity

    data.rename(columns={'Date (dd/mm/yyyy)': 'Date_of_Funding', 'Startup Name': 'Startup_Name',
                         'Industry / Vertical': 'Startup_Industry_Vertical', 'City / Location': 'Startup_Location',
                         'Investors\' Name': 'Investor_Name', 'Investment Type': 'Investment_Type',
                         'Amount (In USD)': 'Amount'}, inplace=True)

    # Removing the rows which had the bad index

    data.drop(data.index[0], inplace=True)

    # Removiong the rows which had the duplicate header names in place

    data = data[data.Startup_Name != "Startup Name"]

    # Adding the column Startup_Country with default value as India because this is the Indian startup list
    data['Startup_Country'] = 'India'

    # Deleting the column name which is not relevant to the major data frame, ie..crunchbase investment

    del data['Investment_Type']

    # Reading the column names

    # Cleaning the crunchbase dataset
    cb_investments = pd.read_csv(r'Python Group-6_Data Focused Python_12thOct2018\crunchbase-investments.csv',
                                 dtype='unicode')

    # Dropping the website columns
    cb_investments = cb_investments.drop(cb_investments.columns[[0, 7]], axis=1)

    cb_investments.rename(columns={'company_name': 'Startup_Name', 'company_country_code': 'Startup_Country',
                                   'company_category_code': 'Startup_Industry_Vertical',
                                   'Industry / Vertical': 'Startup_Industry_Vertical',
                                   'company_region': 'Startup_Location', 'investor_name': 'Investor_Name',
                                   'investor_category_code': 'Investor_Industry_Vertical',
                                   'investor_city': 'Investor_Location', 'raised_amount_usd': 'Amount',
                                   'funded_at': 'Date_of_Funding'}, inplace=True)

    # Deleting the columns which are not required/redundant.

    del cb_investments['company_state_code']

    del cb_investments['investor_country_code']

    del cb_investments['investor_state_code']

    del cb_investments['investor_region']

    del cb_investments['funding_round_type']

    del cb_investments['funded_year']

    del cb_investments['company_city']

    del cb_investments['funded_quarter']

    del cb_investments['funded_month']

    frames = [data, cb_investments]

    result = pd.concat(frames, sort=True)

    result = pd.DataFrame(result)

    result.to_csv(r'Merged_WebScrapping_CrunchBase_Investment.csv',index=False)

    # Removing the NaN values and replacing them with NA

    result.Date_of_Funding.fillna("NA", inplace=True)
    result.Investor_Industry_Vertical.fillna("NA", inplace=True)
    result.Investor_Location.fillna("NA", inplace=True)
    result.Investor_Name.fillna("NA", inplace=True)
    result.Startup_Country.fillna("NA", inplace=True)
    result.Startup_Industry_Vertical.fillna("NA", inplace=True)
    result.Startup_Location.fillna("NA", inplace=True)
    result.Startup_Name.fillna("NA", inplace=True)
    result.Amount.fillna(0, inplace=True)

    # Filtering the rows which do not have any startup name

    result = result[result.Startup_Name != "NA"]

    return result

    pivot_ui(result)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_csv
from sklearn import preprocessing
from sklearn import *
from sklearn.model_selection import *
#df=data.drop('funded_quarter',axis=1, inplace=False)
        
df=data.dropna(inplace=False) 
#df2.dropna(inplace=True) 
for column in df.columns:
    if df[column].dtype == type(object):
        le = preprocessing.LabelEncoder()
        df[column] = le.fit_transform(df[column])
        
y = df.iloc[:,10].tolist() 

df.head()
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(df, y, test_size=0.3)
print(x_train,y_train)
```

```python
# 1 Multinomial Naive Bayes Classifier  
def naive_bayes_classifier(train_x, train_y):  
    from sklearn.naive_bayes import MultinomialNB  
    model = MultinomialNB(alpha=0.01)  
    model.fit(train_x, train_y)  
    return model  
  
# 2 Logistic Regression Classifier  
def logistic_regression_classifier(train_x, train_y):  
    from sklearn.linear_model import LogisticRegression  
    model = LogisticRegression(penalty='l2')  
    model.fit(train_x, train_y)  
    return model  
  
# 3 Random Forest Classifier  
def random_forest_classifier(train_x, train_y):  
    from sklearn.ensemble import RandomForestClassifier  
    model = RandomForestClassifier(n_estimators=2)  
    model.fit(train_x, train_y)  
    return model  
  
# 4 Decision Tree Classifier  
def decision_tree_classifier(train_x, train_y):  
    from sklearn import tree  
    model = tree.DecisionTreeClassifier()  
    model.fit(train_x, train_y)  
    return model  
  
#test_classifiers = ['RF']
test_classifiers = ['NB','LR','RF','DT']  
classifiers = {'NB':naive_bayes_classifier,   
                   'LR':logistic_regression_classifier,  
                   'RF':random_forest_classifier,  
                   'DT':decision_tree_classifier,   
    }  
for classifier in test_classifiers:   
        model = classifiers[classifier](x_train, y_train)  
        predict = model.predict(x_test)  
        print(predict)

is_binary_class = (len(np.unique(y_train)) == 2)

for classifier in test_classifiers:   
        model = classifiers[classifier](x_train, y_train)  
        predict = model.predict(x_test)  
        if is_binary_class:  
            precision = metrics.precision_score(y_test, predict)  
            recall = metrics.recall_score(y_test, predict)  
            print('precision: %.2f%%, recall: %.2f%%' % (100 * precision, 100 * recall))  
        accuracy = metrics.accuracy_score(y_test, predict)  
        print('accuracy: %.2f%%' % (100 * accuracy))
```
